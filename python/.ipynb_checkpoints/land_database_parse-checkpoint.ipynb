```python
capitol_db = db[db["PROP_ID"] == 197003]
```

```python
capitol_geom = None
for elem in capitol_db["the_geom"]:
    capitol_geom = elem
print(capitol_geom)
```

```python
capitol_wkt = load_wkt(capitol_geom)
capitol_cen = capitol_wkt.centroid
```

```python
from math import sin, cos, sqrt, atan2, radians

def lon_lat_distance(point1, point2):
    R = 6370
    lon1 = radians(point1.x)  #insert value
    lat1 = radians(point1.y)
    lon2 = radians(point2.x)
    lat2 = radians(point2.y)

    dlon = lon2 - lon1
    dlat = lat2- lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance
```

```python
dists_to_cap = [0]*len(db["the_geom"])
for idx, geom in enumerate(db["the_geom"]):
    p1 = load_wkt(geom)
    cen = p1.centroid
    dists_to_cap[idx] = lon_lat_distance(cen, capitol_cen)
```

```python
db["dists_to_cap"] = dists_to_cap

db.to_csv("land_database_dists_to_cap.csv",index=False)
```
